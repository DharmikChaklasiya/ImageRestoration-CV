# Preprocessing of all the Data

```python
import os
import time
```

```python
def delete_files(path, prefix):
    for file in os.listdir(path):
        if file.startswith(prefix):
            print(f"Deleting: {file}")
            os.remove(path+"/"+file)

def check_and_delete_missing_images(part, index):
    folder = os.getcwd().replace("\\", "/")
    path = folder+part
    
    if (part[-1] == "1") or (part[-1] == "2"):
        r = 0
    if (part[-1] == "3") or (part[-1] == "4"):
        r = 1 
    if (part[-1] == "5") or (part[-1] == "6"):
        r = 2

    
    print(f"Checking folder: {part}")
    for class_pic in index:
        prefix = f"{r}_{class_pic}_"
        ground_truth = f"{r}_{class_pic}_GT_pose"
        txt = f"{r}_{class_pic}_Parameters"
        picture_name = f"{r}_{class_pic}_pose"
        
        #####################################
        # Check GT Picture
        GT_counter = 0
        for picture in os.listdir(path):
            if picture.startswith(ground_truth):
                GT_counter += 1       
        
        if GT_counter != 1:
            delete_files(path, prefix)
            continue
        ######################################
        # Check txt File
        txt_counter = 0
        for picture in os.listdir(path):
            if picture.startswith(txt):
                txt_counter += 1       
        
        if txt_counter != 1:
            delete_files(path, prefix)
            continue
        #####################################
        counter = 0
        
        for picture in os.listdir(path):
            if picture.startswith(picture_name):
                counter += 1       
                
        if counter != 11:
            delete_files(path, prefix)
            continue
                
                
                            
part1 = "/Part1"
part2 = "/Part2"
part3 = "/Part3"
part4 = "/Part4"
part5 = "/Part5"
part6 = "/Part6"

part1_index = [i for i in range(0,5500)]
part2_index = [i for i in range(5500,11001)]


# test = "/test"
# test_index = [i for i in range(0, 10)]


start = time.time()

check_and_delete_missing_images(part1, part1_index)
check_and_delete_missing_images(part2, part2_index)
check_and_delete_missing_images(part3, part1_index)
check_and_delete_missing_images(part4, part2_index)
check_and_delete_missing_images(part5, part1_index)
check_and_delete_missing_images(part6, part2_index)

end = time.time()
timer = end-start
print(f"Cleaning Done in: {timer} seconds!")
```
